```python
# This cell will load the JSON data we have
import json
import pandas as pd

with open('photodescriptions.json','r') as f:
    photoData = json.load(f)
```

```python
# Loop through photos and append list of labels for each photo to universal list
photoCount = 501
labels = []
for i in range(photoCount):
    for lbl in photoData[i]['shot'+str(i)]:
        labels.append(lbl)
    
# Take the universal list and create a set (only unique values)
labelsSet = set(labels)
print("Total number of unique labels is:", len(labelsSet))

# Create list of tuples with (label, count)
countDict = []
for desc in labelsSet:
    countDict.append((desc, labels.count(desc)))

# Create Pandas Dataframe of data
df = pd.DataFrame.from_records(countDict, columns=['Label', 'Count'], index='Label')
df.sort_values('Count', ascending=False, inplace=True)
df['Percentage'] = (df.Count/photoCount)
df['Percentage'] = pd.Series(["{0:.1f}%".format(val * 100) for val in df['Percentage']], index = df.index)

# Filter data
df = df[df.Count > 20]
df
```
